```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Load the dataset
file_path = "Dry Eyes.csv"  # Update this if needed
df = pd.read_csv(file_path)

# Display the first few rows
print("First 5 rows of the dataset:")
display(df.head())

# Separate numerical and categorical columns
numerical_cols = df.select_dtypes(include=['number']).columns
categorical_cols = df.select_dtypes(exclude=['number']).columns

### 1. Summary Statistics for Numerical Columns ###
print("\nSummary Statistics for Numerical Columns:")
display(df[numerical_cols].describe())

### 2. Correlation Analysis ###
print("\nCorrelation Matrix:")
correlation_matrix = df[numerical_cols].corr()
display(correlation_matrix)

# Heatmap of correlation matrix
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap of Numerical Features")
plt.show()

### 3. Outlier Detection Using IQR ###
print("\nOutlier Detection using IQR Method:")

def detect_outliers(column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers.shape[0]

outlier_counts = {col: detect_outliers(col) for col in numerical_cols}
print("Outliers detected in each numerical column:", outlier_counts)

### 4. Categorical Data Distribution ###
print("\nCategorical Variable Distributions:")

for col in categorical_cols:
    print(f"\n{col} Distribution:")
    print(df[col].value_counts())

    # Plot bar charts for categorical features
    plt.figure(figsize=(6, 4))
    sns.countplot(x=df[col], palette="Set2")
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.xticks(rotation=45)
    plt.show()

### 5. Key Observations from Data ###
print("\nKey Insights from the Data:")

# Insight 1: Average sleep duration and stress level
avg_sleep_duration = df["Sleep duration"].mean()
avg_stress_level = df["Stress level"].mean()
print(f"🔹 The average sleep duration is {avg_sleep_duration:.2f} hours.")
print(f"🔹 The average stress level (on a scale of 1-5) is {avg_stress_level:.2f}.")

# Insight 2: Screen time impact on eye discomfort
screen_time_discomfort = df.groupby("Discomfort Eye-strain")["Average screen time"].mean()
print("\n🔹 Average screen time based on eye discomfort:\n", screen_time_discomfort)

# Insight 3: Smoking vs. Dry Eye Disease
smoking_dry_eye = df.groupby("Smoking")["Dry Eye Disease"].value_counts(normalize=True) * 100
print("\n🔹 Percentage of people with Dry Eye Disease based on smoking status:\n", smoking_dry_eye)

# Insight 4: Stress vs. Sleep Quality
stress_sleep_quality = df.groupby("Stress level")["Sleep quality"].mean()
print("\n🔹 Average sleep quality at different stress levels:\n", stress_sleep_quality)

# Insight 5: Activity level vs. Sleep Duration
activity_sleep = df.groupby(pd.cut(df["Physical activity"], bins=[0, 60, 120, 180]))["Sleep duration"].mean()
print("\n🔹 Average sleep duration based on physical activity levels:\n", activity_sleep)
```

```python
# Import necessary libraries
import pandas as pd
import scipy.stats as stats
import numpy as np

# Load the dataset
file_path = "Dry Eyes.csv"  # Update the path if needed
df = pd.read_csv(file_path)

# ---------------------- 1. T-test: Sleep Duration vs. Stress Level ---------------------- #
print("\n🔹 Hypothesis Test 1: T-test for Sleep Duration & Stress Level")

# Grouping data based on stress levels (1-2 = Low, 4-5 = High)
low_stress = df[df["Stress level"] <= 2]["Sleep duration"]
high_stress = df[df["Stress level"] >= 4]["Sleep duration"]

# Perform independent t-test
t_stat, p_value_ttest = stats.ttest_ind(low_stress, high_stress, equal_var=False)

# Print results
print(f"T-statistic: {t_stat:.4f}, P-value: {p_value_ttest:.4f}")
if p_value_ttest < 0.05:
    print("✅ Significant difference in sleep duration between low and high stress groups.")
else:
    print("❌ No significant difference in sleep duration between low and high stress groups.")

# ---------------------- 2. Chi-Square Test: Smoking vs. Dry Eye Disease ---------------------- #
print("\n🔹 Hypothesis Test 2: Chi-square test for Smoking & Dry Eye Disease")

# Create a contingency table
contingency_table = pd.crosstab(df["Smoking"], df["Dry Eye Disease"])

# Perform chi-square test
chi2_stat, p_value_chi2, _, _ = stats.chi2_contingency(contingency_table)

# Print results
print(f"Chi-square Statistic: {chi2_stat:.4f}, P-value: {p_value_chi2:.4f}")
if p_value_chi2 < 0.05:
    print("✅ Significant association between smoking and Dry Eye Disease.")
else:
    print("❌ No significant association between smoking and Dry Eye Disease.")

# ---------------------- 3. ANOVA: Physical Activity vs. Sleep Duration ---------------------- #
print("\n🔹 Hypothesis Test 3: ANOVA for Physical Activity & Sleep Duration")

# Categorizing Physical Activity Levels
df["Activity Level"] = pd.cut(df["Physical activity"], bins=[0, 60, 120, 180], labels=["Low", "Medium", "High"])

# Get sleep duration data for each activity group
low_activity = df[df["Activity Level"] == "Low"]["Sleep duration"]
medium_activity = df[df["Activity Level"] == "Medium"]["Sleep duration"]
high_activity = df[df["Activity Level"] == "High"]["Sleep duration"]

# Perform one-way ANOVA
anova_stat, p_value_anova = stats.f_oneway(low_activity, medium_activity, high_activity)

# Print results
print(f"ANOVA F-statistic: {anova_stat:.4f}, P-value: {p_value_anova:.4f}")
if p_value_anova < 0.05:
    print("✅ Significant difference in sleep duration among different physical activity levels.")
else:
    print("❌ No significant difference in sleep duration among different physical activity levels.")
```

```python
# Import necessary libraries
import pandas as pd
import statsmodels.api as sm
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
file_path = "Dry Eyes.csv"  # Update if needed
df = pd.read_csv(file_path)

# ---------------------- 1. Probability of Dry Eye Disease ---------------------- #
# Calculate overall probability of Dry Eye Disease
prob_dry_eye = df["Dry Eye Disease"].value_counts(normalize=True)
print("\n🔹 Probability of having Dry Eye Disease:")
print(prob_dry_eye)

# ---------------------- 2. Logistic Regression ---------------------- #
# Define independent variables (habits) and dependent variable (Dry Eye Disease)
# Convert categorical variables to numerical (Yes = 1, No = 0)
df["Smoking"] = df["Smoking"].map({"Y": 1, "N": 0})
df["Alcohol consumption"] = df["Alcohol consumption"].map({"Y": 1, "N": 0})
df["Caffeine consumption"] = df["Caffeine consumption"].map({"Y": 1, "N": 0})
df["Smart device before bed"] = df["Smart device before bed"].map({"Y": 1, "N": 0})
df["Blue-light filter"] = df["Blue-light filter"].map({"Y": 1, "N": 0})
df["Discomfort Eye-strain"] = df["Discomfort Eye-strain"].map({"Y": 1, "N": 0})

# Select features (habits) to analyze
features = ["Smoking", "Alcohol consumption", "Caffeine consumption",
            "Smart device before bed", "Blue-light filter", "Stress level",
            "Average screen time"]

X = df[features]  # Independent variables
X = sm.add_constant(X)  # Add constant for regression
y = df["Dry Eye Disease"].map({"Y": 1, "N": 0})  # Dependent variable (1 = Yes, 0 = No)

# Perform Logistic Regression
logit_model = sm.Logit(y, X)
result = logit_model.fit()

# Print model summary
print("\n🔹 Logistic Regression Results:")
print(result.summary())

# ---------------------- 3. Odds Ratio Analysis ---------------------- #
# Extract odds ratios
odds_ratios = pd.DataFrame({"Feature": X.columns, "Odds Ratio": result.params})
odds_ratios["Odds Ratio"] = odds_ratios["Odds Ratio"].apply(lambda x: round(np.exp(x), 2))

# Sort by impact
odds_ratios = odds_ratios.sort_values(by="Odds Ratio", ascending=False)
print("\n🔹 Odds Ratios (Higher means stronger effect):")
print(odds_ratios)

# ---------------------- 4. Visualizing Impact of Habits ---------------------- #
plt.figure(figsize=(10, 5))
sns.barplot(x="Odds Ratio", y="Feature", data=odds_ratios, palette="coolwarm")
plt.title("Impact of Habits on Dry Eye Disease (Odds Ratios)")
plt.xlabel("Odds Ratio")
plt.ylabel("Habit / Feature")
plt.axvline(1, linestyle="--", color="black", alpha=0.7)
plt.show()
```
